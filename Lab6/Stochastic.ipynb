# Stochastic

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
class RANDU():
    
    def __init__(self, s):
        self.SEED = s
        
    def random(self):
        self.SEED *= 65539
        self.SEED %= 2 ** 31
        return self.SEED / 2 ** 31
```

```python
r = RANDU(99)

for i in range(10):
    print(r.random())
```

```python
nums2 = []
for i in range(254):
    nums2.append(r.random())
    
plt.hist(nums2)
```

```python
x = []
y = []
z = []

for i in range(10000):
    x.append(r.random())
    y.append(r.random())
    z.append(r.random())
```

```python
plt.scatter(x, y, s=1)
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig  = plt.figure()
ax = fig.add_subplot(111, projection = "3d")
ax.view_init(elev=20., azim = 60)
ax.scatter(x, y, z, s=1)
```
