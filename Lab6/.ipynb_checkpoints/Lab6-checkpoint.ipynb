# Chaos Game 1

```python
import numpy as np
import math
import matplotlib.pyplot as plt
%matplotlib inline

def func1(x, y):
    newx = (x/2)
    newy = (y/2)
    
    return newx, newy
    
def func2(x, y):
    newx = (x+1)/2
    newy = (y/2)
    
    return newx, newy
    
def func3(x, y):
    newx = (x + 0.5)/2
    newy = (y + 1)/2
    
    return newx, newy


iterations = 40000

x0 = 0
y0 = 0

points = []

for i in range(iterations):
    rand = np.random.random(1)
    if rand < .34:
        x0, y0 = func1(x0, y0)
    elif .34 <= rand < .67:
        x0, y0 = func2(x0, y0)
    else:
        x0, y0 = func3(x0, y0)
        
    points.append((x0, y0))
```

```python
p = np.array(points)

fig = plt.figure()
ax = fig.add_subplot(111, aspect="equal")

plt.plot(p[:, 0], p[:, 1], "o")
```

# Chaos Game 2

```python
def func1(x, y):
    newx = 0
    newy = 0.16 * y
    
    return newx, newy
    
def func2(x, y):
    newx = (0.85 * x) + (0.04 * y)
    newy = (-0.04 * x) + (0.85 * y) + 1.6
    
    return newx, newy
    
def func3(x, y):
    newx = (0.2 * x) - (0.264 * y)
    newy = (0.23 * x) + (0.22 * y) + 1.6
    
    return newx, newy

def func4(x, y):
    newx = (-0.15 * x) + (0.28 * y)
    newy = (0.26 * x) + (0.24 * y) + 0.44
    
    return newx, newy

iterations = 40000

x0 = 0
y0 = 0

points = []

for i in range(iterations):
    rand = np.random.random(1)
    if rand <= .85:
        x0, y0 = func2(x0, y0)
    elif .85 < rand <= .86:
        x0, y0 = func1(x0, y0)
    elif .86 < rand <= .93:
        x0, y0 = func3(x0, y0)
    else:
        x0, y0 = func4(x0, y0)
        
    points.append((x0, y0))
```

```python
p = np.array(points)

fig = plt.figure()
ax = fig.add_subplot(111, aspect="equal")

plt.plot(p[:, 0], p[:, 1], "o")
```

# Infinite Game

```python
#for payouts from 1 to 1mil find out how many coin tosses it would take
#payouts are what the casino is willing to pay. At first it is infinity.

def infinitegame(trials, maxpayout):

    payouts = []
    
    trialcount = 0

    for i in range(trials):
        trialcount += 1
        playing = True
        flips = 0
        pot = 0
        while playing:
            flip = np.random.random()
            flips += 1
            if flip < .5:
                playing = False


        pot = 2 ** flips
        if pot >= maxpayout:
            pot = maxpayout
            playing = False
        payouts.append(pot)
        
    return payouts

trials = 10000

infpayouts = infinitegame(trials, math.inf)
```

```python
totalpayout = sum(infpayouts) / len(infpayouts)
totalpayout
```

```python
subtrials = 500
maxpayouts = []
for i in range(100):
    maxpayouts.append((i+1)*10000)

payoutList = []

for pay in maxpayouts:
    subtrialPays = []
    for i in range(subtrials):
        game = infinitegame(trials, pay)
        payout = sum(game) / len(game)
        subtrialPays.append(payout)
    totPay = sum(subtrialPays) / subtrials
    payoutList.append(totPay)
```

```python
plt.plot(maxpayouts[:], payoutList[:], "o")
plt.xlabel("Max Payout")
plt.ylabel("Expected Payout (Average)")

print("Overall Average for all Max payouts (1 -> 1000000): " + str(sum(payoutList) / len(maxpayouts)))
```

In the graph above I have plotted the the expected payout against the different limiters of the max payout possible. As you can see the expected value of this game appears to be around $21 or $22 for your winnings as you increase the maximum payout. 

For each run of the game I did 10,000 trials, that helped make the average payout for that game. Each game was set a max payout. I also ran 500 sub trials of the trials. So each payout limit was run 5,000,000 times and the average of all the trials was taken for the graph. I made 100 intervals of 10,000 for the max payouts. 

A fair price to pay the casino would really depend on the payout, but since the larger max payouts are only expected to reach $21 or $22 for the pay out, you wouldn't want to pay more than that for an entry to the game. This means that you should be expecting to, over many games, make your money back.

```python
##Takes the interval from a to b and gives the area under it depending on the func given

#(1/(sqrt(2*pi)))e^((-x^2)/2)

def mc_integrate(func, a, b, c, d):
    
    ##apply the code from the example today. instead of the math.sqrt use the function given to 
    ##determine if the given point is within the range
    goodpts = []
    negpts = []
    badpts = []

    trials = 50000
    good = 0
    neg = 0
    for i in range(trials):
        x = np.random.uniform(a, b)
        y = np.random.uniform(c, d)
        funcy = func(x)
        #if the y value is less than 0 and less than the funcy then it is a bad point
        #if the y value is less than 0 but greater than funcy then it is a good point
        
        if y <= 0 and y >= funcy:
            neg += 1
            goodpts.append((x,y))
        elif y >= 0 and y <= funcy:
            good +=1
            goodpts.append((x,y))
        else:
            badpts.append((x,y))
            
    percGarea = good/trials
    percNarea = neg/trials
    area = (b-a) * (d-c)
    print("Estimation of Integral: " + str((percGarea * area) - (percNarea * area)))
    
    return goodpts, badpts
            
    ##The function will let the user know if the number is below the known x, y point, it is a good point.
    ##So if 1,2 is 
```

```python
def func01(x):
    #(1/(sqrt(2*pi)))e^((-x^2)/2)
    y = (1/(math.sqrt(2 * math.pi))) * (math.e ** ((-x**2) / 2))
    return y

def func02(x):
    y = math.sqrt(math.cos(x)**2 + 1)
    return y

def func03(x):
    y = math.sin(x ** 2)
    return y
```

```python
goodpts, badpts = mc_integrate(func01, -1, 1, 0, .4)

#make a return val for mc_int to output the area of the good points in the integral given

g = np.array(goodpts)
b = np.array(badpts)

fig = plt.figure()
ax = fig.add_subplot(111)
plt.plot(g[:, 0], g[:, 1], "o")
plt.plot(b[:, 0], b[:, 1], "o")
```

```python
goodpts, badpts = mc_integrate(func03, 2, 3, -1, 1)

g = np.array(goodpts)
b = np.array(badpts)

fig = plt.figure()
ax = fig.add_subplot(111)
plt.plot(g[:, 0], g[:, 1], "o")
plt.plot(b[:, 0], b[:, 1], "o")
```

The estimation of the integral is calculated by taking the percentage of the good points above the x axis, contained in the curve, and then multiplied by the total area of the graph subtracted by the area under the x axis.
