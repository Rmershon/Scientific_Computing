```python
import numpy as np
import math
import matplotlib.pyplot as plt
%matplotlib inline

goodpts = []
badpts = []

trials = 500000
good = 0
for i in range(trials):
    x = np.random.random()
    y = np.random.random()
    if math.sqrt(x ** 2 + y ** 2) <= 1:
        good +=1
        goodpts.append((x,y))
    else:
        badpts.append((x,y))
```

```python
4 * good / trials
```

```python
g = np.array(goodpts)
b = np.array(badpts)

fig = plt.figure()
ax = fig.add_subplot(111, aspect="equal")
plt.plot(g[:, 0], g[:, 1], "o")
plt.plot(b[:, 0], b[:, 1], "o")
```
