```python
import pandas
import os

cols = ["X", "Y", "Z", "#"]
newCols = ["X", "Y", "Z"]

Datapoints = []

# Datapoints.append(pandas.read_csv("DATA/default__01.txt", header = None, names = cols, comment = '#', sep = ' '))
# Datapoints.append(pandas.read_csv("DATA/default__02.txt", header = None, names = cols, comment = '#', sep = ' '))
# Datapoints.append(pandas.read_csv("DATA/default__03.txt", header = None, names = cols, comment = '#', sep = ' '))
# Datapoints.append(pandas.read_csv("DATA/default__04.txt", header = None, names = cols, comment = '#', sep = ' '))
# Datapoints.append(pandas.read_csv("DATA/default__05.txt", header = None, names = cols, comment = '#', sep = ' '))


for file in os.listdir('DATA'):
    if file != '.DS_Store':
        
        d = pandas.read_csv('DATA/' + file, header = None, names = cols, comment = '#', sep = ' ', engine='python')
        d.drop(d.columns[3], axis=1, inplace=True)
        Datapoints.append(d)

#p = pandas.read_csv("default__1.txt", header = None, names = cols, comment = '#', sep = ' ')
```

```python
waves = []
temp = []
segs = []

tempx = []
tempy = []
tempz = []

x = []
y = []
z = []

jumps = [12, 13, 14, 15]
for i in range(len(Datapoints)):
    for lab in newCols:
        w = thinkdsp.Wave(Datapoints[i][lab])
        w.normalize()
        waves.append(w)
        t = w.segment(.1, .5)
        if i not in jumps:
            temp.append(t)
            if lab == "X":
                tempx.append(t)
            elif lab  == "Y":
                tempy.append(t)
            else:
                tempz.append(t)
        

for k in range(len(tempx)):
    #print(segment.duration)
    i = 0
    start = 0
    while i < 11:
        x.append(tempx[k].segment(start, (tempx[k].duration/11)))
        y.append(tempy[k].segment(start, (tempy[k].duration/11)))
        z.append(tempz[k].segment(start, (tempz[k].duration/11)))
        
        start += (x[k].duration/11)/2
        i += 1

print(len(x))

#Should be an easier way to do this 

jx = []
jy = []
jz = []

for i in newCols:
    temp = []
    w = thinkdsp.Wave(Datapoints[14][i])
    w.normalize()
    temp.append(w.segment(0, .15))
    temp.append(w.segment(.2, .15))
    temp.append(w.segment(.4, .15))
    temp.append(w.segment(.6, .15))
    temp.append(w.segment(.75, .15))
    temp.append(w.segment(.9, .15))
    temp.append(w.segment(1.05, .13))
    temp.append(w.segment(1.18, .12))
    temp.append(w.segment(1.3, .15))
    temp.append(w.segment(1.45, .1))
    temp.append(w.segment(1.55, .15))
    
    if i == "X":
        jx += temp
    elif i == "Y":
        jy += temp
    else:
        jz += temp
#Do above for the rest
for i in newCols:
    temp = []
    w = thinkdsp.Wave(Datapoints[15][i])
    w.normalize()
    temp.append(w.segment(0, .15))
    temp.append(w.segment(.15, .1))
    temp.append(w.segment(.25, .1))
    temp.append(w.segment(.35, .1))
    temp.append(w.segment(.45, .1))
    temp.append(w.segment(.55, .1))
    temp.append(w.segment(.65, .1))
    temp.append(w.segment(.75, .15))
    temp.append(w.segment(.9, .1))
    temp.append(w.segment(1, .15))
    temp.append(w.segment(1.15, .15))
    
    if i == "X":
        jx += temp
    elif i == "Y":
        jy += temp
    else:
        jz += temp
for i in newCols:
    temp = []
    w = thinkdsp.Wave(Datapoints[12][i])
    w.normalize()
    temp.append(w.segment(0, .075))
    temp.append(w.segment(0.075, 0.075))
    temp.append(w.segment(0.15, .075))
    temp.append(w.segment(0.2, .075))
    temp.append(w.segment(0.275, .075))
    temp.append(w.segment(0.35, .075))
    temp.append(w.segment(0.425, .05))
    temp.append(w.segment(0.475, .075))
    temp.append(w.segment(0.55, .075))
    temp.append(w.segment(0.625, .05))
    temp.append(w.segment(0.675, .075))
    
    if i == "X":
        jx += temp
    elif i == "Y":
        jy += temp
    else:
        jz += temp
            
for i in newCols:
    temp = []
    w = thinkdsp.Wave(Datapoints[13][i])
    w.normalize()
    temp.append(w.segment(0, .075))
    temp.append(w.segment(0.075, .04))
    temp.append(w.segment(0.115, .04))
    temp.append(w.segment(0.155, .04))
    temp.append(w.segment(0.195, .045))
    temp.append(w.segment(0.24, .04))
    temp.append(w.segment(0.28, .04))
    temp.append(w.segment(0.32, .04))
    temp.append(w.segment(0.36, .04))
    temp.append(w.segment(0.4, .04))
    temp.append(w.segment(0.44, .04))
    
    if i == "X":
        jx += temp
    elif i == "Y":
        jy += temp
    else:
        jz += temp
            
            
x += jx
y += jy
z += jz

print(len(x))
```

```python
a = x[len(x) - 1].make_spectrum().peaks()[1:31]
```

```python
spect = []
for i in range(len(x)):
    tx = [val[1] for val in x[i].make_spectrum().peaks()[1:31]]
    ty = [val[1] for val in y[i].make_spectrum().peaks()[1:31]]
    tz = [val[1] for val in z[i].make_spectrum().peaks()[1:31]]
    
    spect.append(tx+ty+tz)
```

```python
spect = np.array(spect)
```

```python
from scipy.cluster.vq import kmeans, vq

kvals = []
dists = []
for k in range(2, 11):
    centroids, distortion = kmeans(spect, k)
    kvals.append(k)
    dists.append(distortion)
    
plt.plot(kvals, dists)
```

You can see from the above graph, showing the clear elbow in the clustering at 4 clusters. This means that 4 is the most effective clustering number. This makes sense becuase we completed 4 activities. 

# PCA

```python
from matplotlib.mlab import PCA
pca = PCA(spect)

fig = plt.figure(figsize=(20, 20))
ax = fig.add_subplot(111)

k = 4

centroids, dist, = kmeans(spect, k)
idx, _ = vq(spect, centroids)

j = 0
words= []
clus0 = []
clus1 = []
clus2 = []
clus3 = []
activity = ["Running", "Walking", "Jumping", "Falling"]
for point in zip(pca.Y[:, 0], pca.Y[:, 1]):
    word = ''
    if j < (len(idx)/4):
        word = activity[0]
    elif (len(idx)/4) <= j < (len(idx)/4)*2:
        word = activity[1]
    elif (len(idx)/4)*2 <= j < (len(idx)/4)*3:
        word = activity[2]
    elif (len(idx)/4)*3 <= j < (len(idx)/4)*4:
        word = activity[3]
        
    if idx[j] == 0:
        clus0.append(word)
    elif idx[j] == 1:
        clus1.append(word)
    elif idx[j] == 2:
        clus2.append(word)
    else:
        clus3.append(word)
    #ax.annotate('%s' % word, xy=point, textcoords='data')
    words.append(word)
    
    j+=1

labels = [max(set(clus0), key=clus0.count), max(set(clus1), key=clus1.count), 
          max(set(clus2), key=clus2.count), max(set(clus3), key=clus3.count)]

for i in range(k):
    plt.plot(pca.Y[idx==i, 0], pca.Y[idx==i, 1], "o", alpha=0.7, label=labels[i]) 
    plt.legend()
    plt.title("PCA: 4 Activities")


```

### Discussion

Above you can see 4 distinct categories. These should represent the 4 activities that we completed. However, if you look in the legend, the most common activity in each is listed there. This shows that walking is not the most popular in any one cluster. The cluster of falling makes sense, they are the most isolated and are clearly their own group. There are obviously going to be some falling in with other clusters, but the majority is its own. 

So cluster 1 might be a tie between the Walking and jumping meaning that there would be all four categories represented.

This would put running and walking next to eachother, and jumping and falling next to eachother. Both of these groupings are between similar activities. So they would make the most sense. 

```python
fig = plt.figure(figsize=(20, 20))
ax = fig.add_subplot(111)

j = 0
words= []
clus0 = []
clus1 = []
clus2 = []
clus3 = []
activity = ["Running", "Walking", "Jumping", "Falling"]
for point in zip(pca.Y[:, 0], pca.Y[:, 1]):
    word = ''
    if j > (len(spect) - (11*4)) - 1:
        word = activity[3]
    elif j <= (11*4) - 1:
        word = activity[0]
    elif (11*4) - 1 < j <= (11*4*2) - 1:
        word = activity[1]
    elif (11*4*2) - 1 < j <= (11*4*3) - 1:
        word = activity[2]
        
    if idx[j] == 0:
        clus0.append(word)
    elif idx[j] == 1:
        clus1.append(word)
    elif idx[j] == 2:
        clus2.append(word)
    elif idx[j] == 3:
        clus3.append(word)
    #ax.annotate('%s' % word, xy=point, textcoords='data')
    words.append(word)
    
    j+=1
    
labels = [max(set(clus0), key=clus0.count), max(set(clus1), key=clus1.count), 
          max(set(clus2), key=clus2.count), max(set(clus3), key=clus3.count)]

j = 0   
for point in zip(pca.Y[:, 0], pca.Y[:, 1]):
    word = ''
    if j > (len(spect) - (11*4)) - 1:
        word = activity[3]
    elif j <= (11*4) - 1:
        word = activity[0]
    elif (11*4) - 1 < j <= (11*4*2) - 1:
        word = activity[1]
    elif (11*4*2) - 1 < j <= (11*4*3) - 1:
        word = activity[2]
    for i in range(k):    
        if point in zip(pca.Y[idx==i, 0], pca.Y[idx==i, 1]):
            if word != labels[i]:
                ax.annotate('%s' % word, xy=point, textcoords='data')
    j+=1

for i in range(k):
    plt.plot(pca.Y[idx==i, 0], pca.Y[idx==i, 1], "o", alpha=0.7, label=labels[i]) 
    plt.legend()
    plt.title("PCA: 4 Activities - minorities labeled")
    
    print(i, labels[i])
```
