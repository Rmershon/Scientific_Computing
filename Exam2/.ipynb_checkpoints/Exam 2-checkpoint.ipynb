## Overfishing

The reproduction rate of fish populations is being outpaced by our appetites and fishing abilities. Without moderation and attention to the aquatic ecosystem, through overfishing we could easy end up in a world without fish.

### Step 1 

```python
trials = 100
days = 1000
fish = 2000

totalPops = []

zeroTimeSteps = []

for i in range(trials):
    fishPop = [fish]
    zero = 0
    for j in range(days):
        tempFish = fishPop[-1]
        
        #documentation says that this is a uniform distribution over the inteverval
        a, b, c = np.random.randint(0, 300 + 1, 3)
        
        tempFish = tempFish - a - b - c
        
        tempFish = tempFish * 1.25
        
        if tempFish - (tempFish % 1) >= 0.5:
            tempFish = math.ceil(tempFish)
        else:
            tempFish = math.floor(tempFish)
            
        if tempFish > fish:
            tempFish = fish
        elif tempFish < 0:
            tempFish = 0
            
        
        fishPop.append(tempFish)
    zeroTimeSteps.append(fishPop.index(0))
    
    totalPops.append(fishPop)

b = np.linspace(0, days, days + 1)
for t in totalPops:
    plt.plot(b, t)
plt.xlim(0, 30)
plt.xlabel("Days")
plt.ylabel("Fish Population")
plt.title("Fish Population - OVERFISHING")

print("The average day that the fish population hits zero is: " + str(sum(zeroTimeSteps)/len(zeroTimeSteps)))
```

As you can see from the above graph much of the trials ran out between 5 and 15 days, with only one trial lasting longer than 20 days. The average, as stated above, was 9.97 days or about 10. This method is clearly not sustainable for the companies or the fish populations. 



### Step 2

```python
trials = 100
days = 1000
fish = 2000

fishMax = 100

totalPops = []

zeroTimeSteps = []
maxIntervals = []

while fishMax <= 250:
    totalPops = []

    for i in range(trials):
        fishPop = [fish]
        zero = 0
        for j in range(days):
            tempFish = fishPop[-1]

            #documentation says that this is a uniform distribution over the inteverval
            a, b, c = np.random.randint(0, 300 + 1, 3)

            compList = [a, b, c]
            for i in range(len(compList)):
                if compList[i] > fishMax:
                    compList[i] = fishMax

            tempFish = (tempFish - compList[0] - compList[1] - compList[2]) * 1.25

            if tempFish - (tempFish % 1) >= 0.5:
                tempFish = math.ceil(tempFish)
            else:
                tempFish = math.floor(tempFish)

            if tempFish > fish:
                tempFish = fish
            elif tempFish < 0:
                tempFish = 0


            fishPop.append(tempFish)

        totalPops.append(fishPop)
    fishMax += 10
    maxIntervals.append(totalPops)

b = np.linspace(0, days, days + 1)
    
fig, ax = plt.subplots(2, 2)

fig.set_size_inches(18.5, 10.5)

ax[0,0].set_title("Fish Population - Max Catch of 100")
ax[0,0].set_xlabel("Days")
ax[0,0].set_ylabel("Fish Population")
for t in maxIntervals[0]:
    ax[0,0].plot(b, t)
ax[0,1].set_title("Fish Population - Max Catch of 150")
ax[0,1].set_xlabel("Days")
ax[0,1].set_ylabel("Fish Population")
for t in maxIntervals[5]:
    ax[0,1].plot(b, t)
ax[1,0].set_title("Fish Population - Max Catch of 200")
ax[1,0].set_xlabel("Days")
ax[1,0].set_ylabel("Fish Population")
for t in maxIntervals[11]:
    ax[1,0].plot(b, t)
ax[1,1].set_title("Fish Population - Max Catch of 250")
ax[1,1].set_xlabel("Days")
ax[1,1].set_ylabel("Fish Population")
for t in maxIntervals[-1]:
    ax[1,1].plot(b, t)

```

Instead there needs to be some limit as to how many fish can be caught per company. I have taken the data the same way as before, recording the population over multiple trials and days per trial. This time there is a max that the companies can fish. 

As you can see in the 4 graphs above, a max catch rate of 200 and 250 looks very similar to the no catch rate. Both will run out of population ver quickly with no sort of sustainablility.

## Bin Packing

### Step 3

```python
FULLCONT = 2000

class Shipment():
    
    def __init__(self, size):
        self.size = size
        self.items  = np.random.randint(100, 2000, size)
        self.contain = np.linspace(0, size - 1, size)
        
    def clone(self):
        c = Shipment(self.size)
        c.items = np.copy(self.items)
        return c
    
    def compContainer(self):
        i = 0 # keeps track of where we are
        j = 0 # keeps track of the container
        while i < self.size:
            vals = 0
            adding = True
            while adding:
                if i < self.size:
                    if vals + self.items[i] <= FULLCONT:
                        vals += self.items[i]
                        self.contain[i] = j
                        i += 1
                    else:
                        adding = False
                else:
                    adding = False
            
            j += 1
                
    
    def containers(self):
        return len(np.unique(self.contain))
    
    def swap(self, i, j):
        t = self.items[i]
        self.items[i] = self.items[j]
        self.items[j] = t
        self.compContainer()
        
    def interchange(self, i, j):
        if j < i:
            t = i
            i = j
            j = t    
        if (j - i) % 2 == 0:
            while i != j:
                self.swap(i, j)
                i += 1
                j -= 1
        else:
            while i < j:
                self.swap(i, j)
                i += 1
                j -= 1
                
def annealing(t, temp, decay, stop):
    cost = []
    badP = []
    while temp > stop:
        bd = t.containers()
        i, j = np.random.randint(0, t.size, 2)
        t.swap(i, j)
        nd = t.containers()
        delta = nd - bd
        if delta >= 0:
            p = np.exp(-delta / temp)
            if np.random.random() > p:
                #bad probability
                badP.append(p)
                t.swap(i, j)
        temp *= decay
        cost.append(t.containers())
    return cost, badP
        
        
def descent(tour):
    cost = []
    finished = False
    while not finished:
        d = tour.containers()
        best_d = d
        best_i = -1
        best_j = -1
        s = tour.size
        finished = True
        for i in range(s):
            for j in range(i + 1, s):
                tour.swap(i, j)
                od = tour.containers()
                if od < best_d:
                    finished = False
                    best_d = od
                    best_i = i
                    best_j = j
                tour.swap(i, j)
        if not finished:
            tour.swap(best_i, best_j)
            cost.append(tour.containers())
    return cost
    
s = Shipment(200)
```

```python
s1 = s.clone()
costList = descent(s1)
fig = plt.figure()
ax = fig.add_subplot(111)
plt.plot([x for x in range(len(costList))], costList)
ax.set_title("Descent - Swap")
ax.set_xlabel("Time (Each Move)")
ax.set_ylabel("Cost")

print("Smallest # of containers: " + str(costList[-1]))
```

```python
s2 = s.clone()
costList2, badP = annealing(s2, 500, .999, .000001)
fig = plt.figure()
ax = fig.add_subplot(111)
plt.plot([x for x in range(len(costList2))], costList2)
ax.set_title("Annealing - Swap")
ax.set_xlabel("Time (Each Move)")
ax.set_ylabel("Cost")

print("Smallest # of containers: " + str(costList2[-1]))
```

The above code used the swapping method, of just swapping items within the list. The descent got fairly close to the annealing, however something that you couldn't tell from just looking at the graphs is that the descent takes much longer, even using less steps. This is interesting because annealing uses up near 20,000 steps and descent doesn't even use 20. This may be because the gradient descent needs more time because it goes through each value in the size of the Shipment. It continues to make minor changes until it cant find a better result than before. This is why it takes so long because for each of these swaps while looking through the list of items, it needs to reorganize the containers each time. And then must keep track of the best intervals throughout, and swap those. So this has the potential of making many more swaps than annealing, meaning it will take much longer. 

Annealing takes one maybe two swaps per round, and it gets its numbers through random ints. So while the amount of 'Time' steps may be greater, the number of swaps it is performing is much less than that of gradient descent, which makes it not only give a better solution but also more efficient with its time. 

The gradient descent had an even descent into its final answer of 128 containers. So at each step it was able to decrease the number of containers needed by one. 

Annealing was able to make some big costs at the begining. And made a lot of guesses but progressively gets better and better. Anealling gave me the best or most 'optimal' solution of 123 containers. 

These were done with just using the swap function, but my swap function acts much differently than the original. My swap funtion takes in an i and j value as before, and then swaps the items with those values. It than looks back at the list of containers and then reorganizes to ensure that each container is as full as it can be. 

The 'compContainer' function ensures that each item is looked at and puts the items together, begining at spot 0, until the weight has reached the max so that they will not exceed the maximum weight. It then moves onto the next item and a new container and the same process is repeated. This is my neighborhood. You can think of it as an assembly line, where items are coming down a conveyer belt and you fill up your containers until you cant fit the next one in and then move on to the next one
