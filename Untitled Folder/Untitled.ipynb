```python
import numpy as np
import random
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from matplotlib.colors import LogNorm
import matplotlib.colors as cc
from matplotlib.colors import ListedColormap

S = 4.294e6 # succeptible
E = 3162 # this number is not right. still not right
I = 10000 # close to the number 
R = 4807

TOTAL = S + E + I + R

EVAPORATE = 0.3
LOWERBOUND = 0.0
DIFFUSION = 2

WHOCOUNT = 10
POPULATION = 400
CUREAMOUNT = 200.0
SIZE = 200
WORDOFMOUTH = 10

betas = [0.2, 0.128, 0.16]
mus = [10, 10.38, 13.31]
alphas = [((2 + 21)/2), 10, 12]

StoE = sum(betas)/len(betas)
EtoI = 1/(sum(mus)/len(mus))
ItoR = 1/(sum(alphas)/len(alphas))

colors = cc.ColorConverter.colors
background = [colors['w']]
pop_cols = [colors['b'],
        colors['y'],
        colors['r'],
        colors['g'],
        colors['c'],
        colors['k']]

bb = ListedColormap(background)
pop_colors = ListedColormap(pop_cols)

def von_neuman_neighbors(i, j, size):
    n = []
    if i > 0:
        n.append((i - 1, j))
    if j > 0:
        n.append((i, j - 1))
    if i < size - 1:
        n.append((i + 1, j))
    if j < size - 1:
        n.append((i, j + 1))
    if i > 0 and j > 0:
        n.append((i - 1,j - 1))
    if i > 0 and j < size - 1:
        n.append((i - 1,j + 1))
    if j > 0 and i < size - 1:
        n.append((i + 1,j - 1))
    if i < size - 1 and j < size - 1:
        n.append((i + 1,j + 1))
    return n

def big_neighbors(i,j,size):
    n = []
    
    return n

class Agent():
    
    def __init__(self, x, y, world, color):
        self.x = x
        self.y = y
        self.world = world
        self.color = color
        
    def wander(self):
        n = von_neuman_neighbors(self.x, self.y, self.world.size)
        self.x, self.y = random.choice(n)
        
    def update(self):
        pass

class WHO(Agent):    
    CURE_AMOUNTSTART = 200.0
    CURE_DECREASE = 10

    states = {"CURING": 5}
    
    def __init__(self, x, y, world, energy=CURE_AMOUNTSTART):
        Agent.__init__(self, x, y, world, 1)
        self.energy = energy
        self.state = WHO.states["CURING"]
        self.color = self.state

    def update(self):
        #self.wander()
        neighbors = von_neuman_neighbors(self.x, self.y, self.world.size)
        max = 0
        count = 0
        i = 0
        for n in neighbors:
            if self.world.cells[n[0], n[1]] > max:
                max = self.world.cells[n[0], n[1]]
                count = i
            i +=1
        if max == 0:
            self.wander()
        else:
            self.x, self.y = neighbors[count]
            
        #Need to have them be attracted to the infected people
        
        for p in self.world.people:
            if ((p.x, p.y)) in neighbors and p.state == People.states["EXPOSED"]:
                p.state = People.states["RECOVERED"]
                p.color = p.state
        
class People(Agent):
    
    states = {"SUSCEPTIBLE": 0, "EXPOSED": 1, "INFECTED": 2, "RECOVERED": 3, "DEAD": 4}
   
    def __init__(self, x, y, world):
        self.clock = 0
        Agent.__init__(self, x, y, world, 0)
        state = np.random.random()
        if state > 0.750:
            self.state = People.states["EXPOSED"]
            random_clock = np.random.randint(0, 51)
            self.clock = random_clock
            
        else:
            self.state = People.states["SUSCEPTIBLE"]
        self.color = self.state
    
    def update(self):
        neighbors = von_neuman_neighbors(self.x, self.y, self.world.size)
        #need a bigger spread for the neighbors. With 200x200 most will never have a true neighbor
        
        if self.state == People.states["SUSCEPTIBLE"]:
            for p in self.world.people:
                if ((p.x, p.y)) in neighbors and p.state == People.states["INFECTED"]:
                    self.state = People.states["EXPOSED"]
                        
            self.wander()
        elif self.state == People.states["EXPOSED"]:
            self.clock += 1
            if self.clock > 15 * 5:
                self.state = People.states["INFECTED"]
            if np.random.random() < 0.6:
                self.wander()
        elif self.state == People.states["RECOVERED"]:
            self.wander()
        elif self.state == People.states["INFECTED"]:
            self.clock +=1
            if self.clock > 15 * 5 + 7 * 5:
                self.state = People.states["DEAD"]
            elif self.clock > 15 * 5 + 4 * 5:
                if np.random.random() < 0.05:
                    self.state = People.states["RECOVERED"]
            if np.random.random() < 0.1:
                self.wander()
                
        self.color = self.state


class CA():
    
    def __init__(self, size):
        self.size = SIZE
        self.cells = np.zeros((SIZE, SIZE))
        self.time = 0
        
        self.people = []
        for p in range(POPULATION):
            i, j = np.random.randint(0, SIZE, 2)
            self.people.append(People(i, j, self))

        self.who = []
        for s in range(WHOCOUNT):
            i, j = np.random.randint(0, SIZE, 2)
            self.who.append(WHO(i, j, self))
            
        self.s = []
        self.e = []
        self.i = []
        self.r = []
        self.d = []
        
        
    def image_setup(self):
        self.plt = plt.imshow(self.cells, interpolation='nearest', 
                            origin='bottom', vmin=0, cmap=bb)
        ax = []
        ay = []
        ac = []
        for a in self.people + self.who:
            ax.append(a.x)
            ay.append(a.y)
            ac.append(a.color)
        self.ant_plot = plt.scatter(ax, ay, c=ac, s=20,vmin=0, vmax=5, cmap=pop_colors)

    def update(self):
        newcells = np.zeros((self.size, self.size))
#         for a in self.people + self.who:
#             a.update()
        for i in range(self.size):
            for j in range(self.size):
                neighbors = von_neuman_neighbors(i, j, self.size)               
                otherP = self.cells[i,j]
                for a, b in neighbors:
                    otherP += self.cells[a,b]
                otherP /= len(neighbors) + 1
                newcells[i,j] = (1 - EVAPORATE) * (self.cells[i,j] + (DIFFUSION * (otherP - self.cells[i,j])))
                if newcells[i,j] < LOWERBOUND:
                    newcells[i,j] = LOWERBOUND
        
        self.cells = newcells
        
        states = []
        
        for a in self.people + self.who:
            if a.state == People.states["INFECTED"]:
                self.cells[a.x, a.y] = 10
            if a.state == People.states["EXPOSED"] and np.random.random() > 0.2:
                self.cells[a.x, a.y] = 10
            if a.state == People.states["RECOVERED"]:
                self.cells[a.x, a.y]
            a.update()
            states.append(a.state)
            
        states = np.array(states)
        self.s.append(int(np.count_nonzero(states == 0)))
        self.e.append(int(np.count_nonzero(states == 1)))
        self.i.append(int(np.count_nonzero(states == 2)))
        self.r.append(int(np.count_nonzero(states == 3)))
        self.d.append(int(np.count_nonzero(states == 4)))

        self.time += 1
        
    def plot(self):
        self.plt.set_data(np.transpose(self.cells))
        ap = []
        ac = []
        for a in self.people:
            ap.append((a.x, a.y))
            ac.append(a.color)
        for f in self.who:
            ap.append((f.x, f.y))
            ac.append(f.color)
        self.ant_plot.set_offsets(ap)
        ac = np.array(ac)
        self.ant_plot.set_array(ac)
            

        return self.plt, self.ant_plot
        

fig, ax = plt.subplots()
ax.set_ylim(-1, SIZE)
ax.set_xlim(-1, SIZE)

ca = CA(SIZE)
ca.image_setup()
#ca.random_setup(100)

       
def update(data):
    ca.update()
    return ca.plot(), 

def data_gen():
    while True: yield 1

ani = animation.FuncAnimation(fig, update, data_gen, blit=False, interval=10)
plt.show()
```
