```python
from urllib.request import urlopen
import urllib.request as http
import json
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
def zipFinder(zipCode):
    try:
        r = urlopen("http://ziplocate.us/api/v1/" + zipCode)
        data = r.read().decode()
        r.close()
        jsonObj = json.loads(data)
    except http.HTTPError as err:
        if err.code == 404:
            return None
        else:
            raise
        
    
    return jsonObj["lat"], jsonObj["lng"], jsonObj["zip"]
```

```python
zipFinder("78735")
```

```python
lats = []
lngs = []

for num in range(71601, 72959):
    tempObj = zipFinder(str(num))
    
    if tempObj != None:
        lats.append(tempObj[0])
        lngs.append(tempObj[1])
    
    
```

```python
print(lats)
print(lngs)
```

```python
plt.scatter(lngs, lats)
plt.axes().set_aspect('equal')
plt.ylim([32, 38])
plt.xlim([-95, -89])
```
